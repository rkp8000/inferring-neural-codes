```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

import pandas as pd

cc = np.concatenate

FPS = 30.03
DT = 1/FPS
```

```python
df_behav = pd.read_csv('data/simple/c_song_f_behav.csv')
cols = df_behav.columns
```

```python
# split big df into dfs for individual trials
ntr = np.max(df_behav.ID) + 1
df_tr_0 = [df_behav[df_behav.ID == i] for i in range(1)][0]
df_behav = None
```

```python
t = np.arange(0, 370, DT)
i = ((10 <= t)&(t < 70))
```

```python
# sine block trial (10 sec Q, 60 sec S, 300 sec P)
df_s = pd.DataFrame(columns=cols)
df_s['T'] = t
df_s['ID'] = 0
df_s['FRAME'] = np.arange(len(t))
df_s['Q'] = (~i).astype(int)
df_s['S'] = i.astype(int)
df_s['P'] = 0
df_s['F'] = 0

for kbhv in ['FFV', 'FFA', 'FLS', 'FRS', 'FMFV', 'FMLS']:
    df_s[kbhv] = np.random.randn(len(t))
    
# pulse block trial (10 sec Q, 60 sec S, 300 sec P)
df_p = pd.DataFrame(columns=cols)
df_p['T'] = t
df_p['ID'] = 1
df_p['FRAME'] = np.arange(len(t))
df_p['Q'] = (~i).astype(int)
df_p['P'] = i.astype(int)
df_p['S'] = 0
df_p['F'] = 0

for kbhv in ['FFV', 'FFA', 'FLS', 'FRS', 'FMFV', 'FMLS']:
    df_p[kbhv] = np.random.randn(len(t))
```

```python
df_cc = pd.concat([df_s, df_p], ignore_index=True)
```
