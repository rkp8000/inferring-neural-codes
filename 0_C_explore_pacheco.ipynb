```python
stims = [None] + [stim[0] for stim in d['stimuli_name_string'].flatten()]
n_stim = len(stims)-1
stims
```

```python
bds = [None]

for istim in range(1, 4):
    segs, bds_seg = get_seg(stim_index==istim, min_gap=1)
    bds.append(bds_seg[0])
    
print(bds)
```

```python
flies = np.unique(d['fly_index_per_roi'])
mflys = [(d['fly_index_per_roi'] == fly).flatten() for fly in flies]

expts = []

for fly, mfly in zip(flies, mflys):
    print(f'Fly {fly}: {mfly.sum()} rois')
    expts.append(med_rsp[mfly, :])
    
print()

for fly, expt in zip(flies, expts):
    print(f'Fly {fly}: {expt.shape}')
```

```python
ifly = 0
t = -10 + np.arange(len(d['timestamps'].flatten()))*.5

fig, ax = plt.subplots(1, 1, figsize=(15, 17), tight_layout=True)
for ctr in range(30):
    ax.axhline(4*ctr, lw=1, ls='--', c='gray')
    ax.plot(t, -expts[ifly][ctr, :] + 4*ctr, lw=3)

colors = [None, 'r', 'b', 'gray', 'm']
for cstim, stim in enumerate(stims[1:], 1):
    start, end = bds[cstim]
    ax.axvspan(t[start], t[end], color=colors[cstim], alpha=.2)
    
ax.set_ylim(4*30, -4)
ax.set_xlabel('Time (s)')
ax.set_ylabel('ROI index')
ax.set_title(f'Fly {flies[ifly]}')

set_font_size(ax, 16)
```

```python
n_t_stim = len(t_orig)/n_stim
tmp = np.arange(len(t_orig))
mstims = [None] + [(i*n_t_stim <= tmp)&(tmp < (i+1)*n_t_stim) for i in range(n_stim)]
```

Do PCA

```python
for ifly in range(5):
    expt = expts[ifly]
    n_roi, n_t = expt.shape
    n_comp = min(n_roi, n_t)

    pca = decomposition.PCA(n_components=n_comp)
    pca.fit(expts[ifly].T)

    pca.components_.shape  # one row per component
    fig, axs = plt.subplots(1, 2, figsize=(10, 4), tight_layout=True)

    axs[0].scatter(range(n_comp), pca.explained_variance_ratio_)
    axs[0].set_xlabel('Component')
    axs[0].set_ylabel('Frac var explained')

    axs[1].scatter(range(n_comp), pca.explained_variance_ratio_.cumsum())
    axs[1].set_xlabel('Kept components')
    axs[1].set_ylabel('Frac var explained')

    for ax in axs:
        ax.set_ylim(-.05, 1.05)
        ax.grid()
        ax.set_title(f'Fly {flies[ifly]}')
    
    set_font_size(axs, 16)
        
    pc_0, pc_1, pc_2, pc_3 = pca.components_[:4, :]
        
    fig, axs = plt.subplots(2, 1, figsize=(12, 7), tight_layout=True)
    axs[0].plot(t, pc_0 @ expt, c='k', lw=3)
    axs[0].plot(t, pc_1 @ expt, c='g', lw=3)
    axs[0].plot(t, pc_2 @ expt, c='y', lw=3)
    axs[0].plot(t, pc_3 @ expt, c='orange', lw=3)

    axs[0].set_xlabel('Time (s)')
    axs[0].set_ylabel('Projection onto PC')
    axs[0].set_title(f'Fly {flies[ifly]}')
    axs[0].legend(['PC 1', 'PC 2', 'PC 3', 'PC 4'], loc='upper left')
    
    axs[1].plot(t, expt.mean(axis=0), c='k', lw=3, ls='--')
    axs[1].set_xlabel('Time (s)')
    axs[1].set_ylabel('Mean response')
    axs[1].set_title('Mean response')
    
    colors = [None, 'r', 'b', 'k', 'm']
    for ax in axs:
        for cstim, stim in enumerate(stims[1:], 1):
            start, end = bds[cstim]
            ax.axvspan(t[start], t[end], color=colors[cstim], alpha=.2)
        ax.grid()
        set_font_size(ax, 16)
        
    fig, axs = plt.subplots(1, 3, figsize=(15, 5), tight_layout=True)
    
    for istim in range(1, 4):
        mstim = mstims[istim]
        axs[0].plot((pc_0 @ expt)[mstim], (pc_1 @ expt)[mstim], c=colors[istim], lw=3)
        axs[1].plot((pc_0 @ expt)[mstim], (pc_2 @ expt)[mstim], c=colors[istim], lw=3)
        axs[2].plot((pc_1 @ expt)[mstim], (pc_2 @ expt)[mstim], c=colors[istim], lw=3)
        
    axs[0].set_xlabel('PC 1')
    axs[0].set_ylabel('PC 2')
    axs[1].set_xlabel('PC 1')
    axs[1].set_ylabel('PC 3')
    axs[2].set_xlabel('PC 2')
    axs[2].set_ylabel('PC 3')
    
    for ax in axs:
        ax.set_title(f'Fly {flies[ifly]}')
    set_font_size(axs, 16)
    
    fig = plt.figure(figsize=(15, 7), tight_layout=True)
    axs = np.array([fig.add_subplot(121, projection='3d'), fig.add_subplot(122, projection='3d')])
    
    for istim in range(1, 4):
        mstim = mstims[istim]
        axs[0].plot((pc_0 @ expt)[mstim], (pc_1 @ expt)[mstim], (pc_2 @ expt)[mstim], c=colors[istim], lw=3)
        axs[1].plot((pc_1 @ expt)[mstim], (pc_2 @ expt)[mstim], (pc_3 @ expt)[mstim], c=colors[istim], lw=3)
        
    axs[0].set_xlabel('PC 1')
    axs[0].set_ylabel('PC 2')
    axs[0].set_zlabel('PC 3')
    
    axs[1].set_xlabel('PC 2')
    axs[1].set_ylabel('PC 3')
    axs[1].set_zlabel('PC 4')
    
    for ax in axs:
        ax.set_title(f'Fly {flies[ifly]}')
    set_font_size(axs, 16)
```
