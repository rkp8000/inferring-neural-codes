Plot and manually segment pacheco nat song segment

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import loadmat
```

```python
data = loadmat('data/auditory_stimuli_vectors.mat')
```

```python
nat_stim = data['stimuli'][0][0][3]
bds_pulse = [(3.5, 3.625), (4.05, 4.18), (4.68, 4.75), (5.15, 5.2), (5.58, 5.7), (6.02, 6.85), (7.69, 8.46), (9.16, 9.71)]
```

```python
fig, ax = plt.subplots(1, 1, figsize=(20, 3), tight_layout=True)
t = np.arange(len(nat_stim))/10000
ax.plot(t, nat_stim)
for lb, ub in bds_pulse:
    mt = (lb <= t)&(t < ub)
    ax.plot(t[mt], nat_stim[mt], c='r')
ax.set_xticks(np.arange(0, 10.01, .25))
ax.grid()
```
