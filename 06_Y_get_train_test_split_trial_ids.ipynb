```python
SEED = 0
PTRAIN = .8
NSPLIT = 30
```

```python
# SPECIFY STRAINS USED
STRAINS = ['NM91', 'ZH23']
STRAIN_KEY = '_'.join(STRAINS).lower()

FSTRAIN = 'data/simple/strains.csv'
MSTRAINS = [(pd.read_csv(FSTRAIN)['STRAIN'] == strain) for strain in STRAINS]
MSTRAIN = np.any(MSTRAINS, axis=0)
ISTRAIN = MSTRAIN.nonzero()[0]

NTR = len(ISTRAIN)
NTRAIN = int(round(PTRAIN*NTR))
NTEST = NTR - NTRAIN
```

```python
np.random.seed(SEED)

for csplit in range(NSPLIT):
    irnd = np.random.permutation(NTR)
    itr_train = irnd[:NTRAIN]
    itr_test = irnd[NTRAIN:]
    
    print(f'SPLIT {csplit}')
    print('Test trials:', list(zip(range(NTEST), itr_test)))
```
