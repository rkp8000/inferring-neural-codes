```python
T = 10
DT = .0000001

t = np.arange(-T, T, DT)
f = 1+np.linspace(0, 30, len(t))

osc = np.sin(2*np.pi*f*t)
# s = (osc > 0).astype(float)

fig, axs = plt.subplots(3, 1, figsize=(15, 6), tight_layout=True)
axs[0].plot(t, f)
axs[0].set_ylabel('f')
axs[1].plot(t, osc)
axs[1].set_xlabel('t')
# axs[2].plot(t, s)
```
