# Pacheco et al NA fit param distributions

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from disp import get_line, set_plot

FPS = 2
DT = 1/FPS
T = np.arange(61)*DT - 10
T_COLS = [f'IT_{it}' for it in range(61)]
```

```python
df_ma_param = pd.read_csv('data/simple/neur/pacheco_ma_fit_param.csv')
df_ma_pred = pd.read_csv('data/simple/neur/pacheco_ma_fit_pred.csv')

df_data = pd.read_csv('data/simple/neur/pacheco.csv')

mout = np.zeros(len(df_ma_param['TAU_R']), dtype=bool)
```

```python
df_data
```

```python
# multiplicative adaptation fit param distributions
fig, axs = plt.subplots(1, 4, figsize=(10, 2.5), tight_layout=True)
axs[0].hist(1/df_ma_param['TAU_R'][~mout], bins=np.linspace(0, 3, 30))
axs[0].set_xlabel(r'$1/\tau_{int}$ (1/s)')
axs[1].hist(1/df_ma_param['TAU_A'][~mout], bins=np.linspace(0, 3, 30))
axs[1].set_xlabel(r'$1/\tau_a$ (1/s)')
axs[2].hist(df_ma_param['X_S'][~mout], bins=np.linspace(-20, 20, 30))
axs[2].set_xlabel(r'$x_s$')
axs[3].hist(df_ma_param['X_P'][~mout], bins=np.linspace(-20, 20, 30))
axs[3].set_xlabel(r'$x_p$')

axs[0].set_ylabel('# ROIs')

for ax in axs:
    set_plot(ax, font_size=14)

fig.savefig('supp/s1_na_params_pacheco.svg')
```

```python
mfasi = (2 <= df_ma_param['TAU_R']) & (df_ma_param['TAU_R'] < 120) & (.1 <= df_ma_param['TAU_A']) & (df_ma_param['TAU_A'] < 2)

fig, ax = plt.subplots(1, 1, figsize=(3, 3), tight_layout=True)
ax.scatter(1/df_ma_param['TAU_R'][mfasi], 1/df_ma_param['TAU_A'][mfasi], c='g', s=5)
ax.scatter(1/df_ma_param['TAU_R'][~mfasi], 1/df_ma_param['TAU_A'][~mfasi], c='k', s=5)
ax.legend(['Fast-ad-slow-int'])
set_plot(ax, x_label=r'1/$\tau_{int}$ (1/s)', y_label=r'1/$\tau_a$ (1/s)')


fig.savefig('supp/s7_na_params_pacheco_fast_ad_slow_int.svg')
```

```python
print('Total recordings', len(mfasi))
print('Number of FASI responses', mfasi.sum())
print('Fraction', mfasi.sum()/len(mfasi))
```
