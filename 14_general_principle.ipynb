```python
import matplotlib.pyplot as plt
import numpy as np

from disp import plot_b, get_spaced_colors, set_plot

cc = np.concatenate
```

```python
K = 20
NSEG = 60
TAU = 2

colors = get_spaced_colors('jet', K, .3)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 2), tight_layout=True)

seq = cc([np.repeat(np.random.randint(0, K), np.round(np.random.exponential(TAU))) for _ in range(NSEG)])
t = np.arange(len(seq))

plot_b(ax, t, seq, [0, t.max(), 0, 1], c=colors)
set_plot(ax, y_lim=(-1, 5))

fig.savefig('ms/fig_6_string_xpl.svg')
```

```python
# advection diffusion schematic
SEED = np.random.randint(0, 2**32-1)
SEED = 3411920491
np.random.seed(SEED)
print(SEED)
T = 300
x = (np.random.randn(T) + .1).cumsum()
y = np.random.randn(T).cumsum()
z = np.random.randn(T).cumsum()
X_LIM = (0, 50)
YZ_LIM = (-30, 30)

fig = plt.figure(figsize=(10, 10), tight_layout=True)
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, c='k', lw=2)
set_plot(ax, x_lim=X_LIM, y_lim=YZ_LIM, x_label='X', y_label='Y')
ax.set_zlim(*YZ_LIM)

fig.savefig('ms/fig_6_neural_trj.svg')
```

```python
w = np.random.randn(3)

b = np.array([x, y, z]).T@w

fig, ax = plt.subplots(1, 1, figsize=(6, 3), tight_layout=True)
ax.plot(b, color='orange', lw=2)

fig.savefig('ms/fig_6_behav_xpl.svg')
```
